<h3>RBG to grayscale using <code>tf.image.rgb_to_grayscale</code></h3>
<p>In this sample notebook you will see how converting an RBG image to gray scale works on an image</p>

```
%tensorflow_version 2.x
import tensorflow as tf 
import numpy as np
from tensorflow.keras.preprocessing import image as IMG 
from skimage import io
```

<p>Since am working in a colab environment, i will use my drive to store the converted image.
If you are using your local environment, you can skip this.</p>

```
from google.colab import drive
drive.mount('/content/drive')
```

```
#url to sample image
image = 'https://wonderfulengineering.com/wp-content/uploads/2016/07/This-Is-The-Reason-Why-Google-Is-Open-Sourcing-Some-Of-Its-Most-Important-Technology_Image-0.jpg'

#getting the image from the url
read_img = io.imread(image)

io.imshow(read_img)
```

print shape of the image to be converted to grayscale to ensure that the last dimension value is 3

```
print(read_img.shape)
```

Convert the image to grayscale. This is where the image is converted to gray scale

```
gray_image = tf.image.rgb_to_grayscale(read_img, name=None)
```

printing the shape of the gray_image. We will see that the last dimension value is now 1

```
 print(gray_image.shape)
```

Converting the image into numpy array inorder to save it

```
np_array =  IMG.img_to_array(gray_image)
```

```
#path to where the image will be saved
fname = '/content/drive/My Drive/gray_image.jpg'

#save image
io.imsave(fname, np_array)
```

```
#reading the saved image
img = fname

#showing the the image
io.imshow(img)
```

<p>From the above image we can see that our image was converted successfully to gray scale. Thank you for going through this sample usage of tf.image.rgb_to_grayscale on an image</p>
